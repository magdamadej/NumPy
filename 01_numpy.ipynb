### Podstawowe operacje na tablicach

```python
A = np.array([3, 1, 4, 2])
B = np.array([3, -1, 3, 2])
print(A)
print(B)
```

```python
# dodawanie - taka sama ilość elementów
A + B
```

```python
# odejmowanie
A - B
```

```python
A * B
```

```python
A / B
```

```python
A + 3
```

```python
2 * A
```

```python
A + 3*B
```

```python
np.add(A, B)
```

```python
np.subtract(A, B)
```

```python
np.multiply(A, B)
```

```python
X = np.array([[1, 3], [-2, 0]])
Y = np.array([[6, 0], [-1, 2]])
print(X, '\n')
print(Y)
```

```python
# mnożenie macierzy
np.dot(X, Y)
# lub X.dot(Y)
# lub X @ Y
```

### Iteracja po tablicach

```python
# iteracja po wierszach
for row in A:
  print(row)
```

```python
for row in A:
  print(row[:3])
```

```python
# iteracja po każdym elemencie
for item in A.flat:
  print(item)
```
